```python
import pandas as pd

df = pd.read_csv('../1_data/sales_data_sample.csv', encoding='latin1')

print('Filas y columnas:', df.shape)
df.head()
```

```python
print(df.dtypes)
```

```python
print('Valores nulos por columna:')
print(df.isnull().sum())
```

```python
df.describe()
```

```python
duplicados = df.duplicated().sum()
print(f'Filas duplicadas: {duplicados}')
```

```python
print('PaÃ­ses Ãºnicos:')
print(df['COUNTRY'].unique())

print(f'\nTotal productos Ãºnicos: {df["PRODUCTCODE"].nunique()}')
print(f'Total Ã³rdenes Ãºnicas: {df["ORDERNUMBER"].nunique()}')
```

```python
columnas_relevantes = [
    'ORDERNUMBER',
    'ORDERDATE',
    'PRODUCTCODE',
    'QUANTITYORDERED',
    'PRICEEACH',
    'SALES',
    'COUNTRY',
    'CUSTOMERNAME',
    'PRODUCTLINE',
    'DEALSIZE',
    'STATUS'
]

df_limpio = df[columnas_relevantes].copy()
print('Columnas seleccionadas:', df_limpio.shape)
df_limpio.head()
```

```python
df_limpio['ORDERDATE'] = pd.to_datetime(df_limpio['ORDERDATE'])

print('Tipo de dato ORDERDATE:', df_limpio['ORDERDATE'].dtype)
print('Fecha mÃ¡s antigua:', df_limpio['ORDERDATE'].min())
print('Fecha mÃ¡s reciente:', df_limpio['ORDERDATE'].max())
```

```python
df_limpio.columns = df_limpio.columns.str.lower()

print('Columnas finales:')
print(df_limpio.columns.tolist())
```

```python
# Extraer productos Ãºnicos con su lÃ­nea de producto
productos = df_limpio[['productcode', 'productline']].drop_duplicates().reset_index(drop=True)

# Crear el ID numÃ©rico
productos.insert(0, 'producto_id', range(1, len(productos) + 1))

# Renombrar columnas
productos.columns = ['producto_id', 'producto_nombre', 'categoria']

print('Total productos:', len(productos))
productos.head(10)
```

```python
# Unir el producto_id a las ventas
df_limpio = df_limpio.merge(
    productos[['producto_id', 'producto_nombre']],
    left_on='productcode',
    right_on='producto_nombre',
    how='left'
)

# Verificar
print('Columnas actualizadas:')
print(df_limpio.columns.tolist())
df_limpio[['ordernumber', 'productcode', 'producto_id']].head()
```

```python
# Eliminar columnas que ya no necesitamos
df_limpio = df_limpio.drop(columns=['productcode', 'producto_nombre'])

print('Columnas finales de ventas:')
print(df_limpio.columns.tolist())
df_limpio.head()
```

```python
print('Nulos en producto_id:', df_limpio['producto_id'].isnull().sum())
print('Total filas:', len(df_limpio))
```

```python
# Guardar ventas con producto_id
df_limpio.to_csv('../4_output/ventas_limpio.csv', index=False)

# Guardar tabla de productos
productos.to_csv('../4_output/productos.csv', index=False)

print('Archivos guardados en 4_output/ â')
```
